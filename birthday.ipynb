## Understanding The Problem
* If there are n people in a room, how large does n need to be for at least 50% chance of at least two people sharing a birthday?
* The problem is easier to compute by computing the probability that nobody shares a birthday and take one minus that probability using the compliment rule.
* Compliment Rule: $P(A^c) = 1-P(A)$
* So for this problem that translates to: $P(\geq 2shares)=1-P(noShares)$
* To calculate the probability of no one sharing a birthday : $P(noOneShares)=\frac{365!}{(365-n)!\times365^n}$

```python
import matplotlib.pyplot as plt
import math
from decimal import Decimal, getcontext
```

## Solving The Problem


*Compute:  $P(noOneShares)=\frac{365!}{(365-n)!\times365^n}$*

```python
getcontext().prec = 10
def nobody_shares(n):
    total_no_unique_birthdays = Decimal(math.factorial(365)) / Decimal(math.factorial(365-n))
    total_no_choices = Decimal(365) ** n
    return total_no_unique_birthdays * (1/total_no_choices)
```

*Loop For 100 People:*

```python
data = {}
n = 1
while True:
    probability = 1 - nobody_shares(n)
    if n > 100:
        break
    data[n] = probability
    n += 1
print(data)
```

## Plot The Data
##### Red Line shows how many (n) people need to be in a room for at least 50% chance of at least 2 people sharing a brithday

```python
x = list(data.keys())
y = list(data.values())

plt.plot(x,y)
plt.title("Birthday Problem")
plt.xlabel("number of people (n)")
plt.ylabel("P(>= 2 shares")
plt.grid(True)

#Plot Red Line Where Probability = 0.5
plt.axhline(y=0.5, color='r')

plt.show()
```
